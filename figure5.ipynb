```python
import matplotlib.pyplot as plt
import numpy  as np
import xarray as xr
```

GLIDER

```python
# EXTRACTED FROM CODE FIGURE 3

rmsd_gli_swot= 0.622
rmsd_gli_duacs = 0.822

percent_gli = 100*(rmsd_gli_duacs - rmsd_gli_swot)/rmsd_gli_duacs

err_gli_swot = 0.06
err_gli_duacs = 0.1

percent_gli
```

ADCP

```python
# EXTRACTED FROM CODE adcp_compute_rmsd_all

rmsd_adcp_swot = 8.88
rmsd_adcp_duacs = 13.88

percent_adcp = 100*(rmsd_adcp_duacs - rmsd_adcp_swot)/rmsd_adcp_duacs

err_adcp_swot = 0.4
err_adcp_duacs = 0.3

percent_adcp
```

```python
# Grouping the data
rmsd = [rmsd_gli_duacs, rmsd_adcp_duacs]
rmsd_swot = [rmsd_gli_swot, rmsd_adcp_swot]
percent_improv = [percent_gli, percent_adcp]

labels = ['GLI [cm]', 'ADCP [cm/s]']
```

SVPB drifters

```python
rmsd_swot_modul = 7.4
rmsd_duacs_modul = 10.76

rmsd_swot_dir = 34.16
rmsd_duacs_dir = 30.63

percent_improv_modul = 100*(rmsd_duacs_modul - rmsd_swot_modul)/rmsd_duacs_modul
percent_improv_dir = 100*(rmsd_duacs_dir - rmsd_swot_dir)/rmsd_duacs_dir

err_modul_swot = 0.5
err_modul_duacs = 0.3

err_dir_swot = 3.2
err_dir_duacs = 1.8
```

### FIGURE

```python
fig = plt.figure(figsize=(8.5, 5))
gs = fig.add_gridspec(1, 3, width_ratios=[1, 1, 2.1], wspace=0.4)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[0, 2])
x = np.array([0])
width = 0.25

# Subplot 1
ax1.bar(x + 0.05, rmsd[0], width / 3.2, label='vs DUACS', color='darkcyan', zorder=2, alpha=0.9)
ax1.errorbar(x + 0.05, rmsd[0], yerr=err_gli_duacs, fmt='o', color='yellowgreen', zorder=3, capsize=3)
ax1.bar(x + 0.132, rmsd_swot[0], width / 3.2, label='vs SWOT', color='skyblue', zorder=2, alpha=0.9)
ax1.errorbar(x + 0.133, rmsd_swot[0], yerr=err_gli_swot, fmt='o', color='slateblue', zorder=3, capsize=3)

ax1.set_ylabel('RMSD [cm]', fontsize=10)
ax1.set_xticks([0.09])
ax1.set_xticklabels(['Glider dynamic height'], fontsize=10)
ax1.yaxis.set_ticks(np.arange(0, 1.4, 0.2))
ax1.yaxis.grid(True)
ax1.tick_params(axis='y', labelsize=10)
ax1.set_ylim([0, 1])

bar_reduction = 100 * (rmsd[0] - rmsd_swot[0]) / rmsd[0]
ax1.text(0.09, 0.1, f'{bar_reduction:.0f}% RMSD \n reduction', ha='center', va='center',
         fontsize=9, fontweight='bold',
         bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='black', alpha = 0.9))

# Subplot 2
ax2.bar(x + 0.07, rmsd[1], width * 1.3, label='vs DUACS', color='darkcyan', zorder=2, alpha=0.9)
ax2.errorbar(x + 0.07, rmsd[1], yerr=err_adcp_duacs, fmt='o', color='yellowgreen', zorder=3, capsize=3)
ax2.bar(x + 0.41, rmsd_swot[1], width * 1.3, label='vs SWOT', color='skyblue', zorder=2, alpha=0.9)
ax2.errorbar(x + 0.4, rmsd_swot[1], yerr=err_adcp_swot, fmt='o', color='slateblue', zorder=3, capsize=3)

ax2.set_ylabel('RMSD [cm/s]', fontsize=10)
ax2.set_xticks([0.24])
ax2.set_xticklabels(['ADCP cross-track vel.'], fontsize=10)
ax2.yaxis.set_ticks(np.arange(0, 19, 3))
ax2.yaxis.grid(True)
ax2.tick_params(axis='y', labelsize=10)
ax2.set_ylim([0, 15])

bar_reduction = 100 * (rmsd[1] - rmsd_swot[1]) / rmsd[1]
ax2.text(0.24, 1.5, f'{bar_reduction:.0f}% RMSD \n reduction', ha='center', va='center',
         fontsize=9, fontweight='bold',
         bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='black', alpha = 0.9))

# Subplot 3
ax3_right = ax3.twinx()
ax3.bar(x + 0.03, rmsd_duacs_modul, width * 1.3, color='darkcyan', zorder=2, label='vs DUACS')
ax3.errorbar(x + 0.03, rmsd_duacs_modul, yerr=err_modul_swot, fmt='o', color='yellowgreen', zorder=3, capsize=3)
ax3.bar(x + 0.37, rmsd_swot_modul, width * 1.3, color='skyblue', zorder=2, label='vs SWOT')
ax3.errorbar(x + 0.37, rmsd_swot_modul, yerr=err_modul_duacs, fmt='o', color='slateblue', zorder=3, capsize=3)

ax3.set_ylabel('RMSD [cm/s]', fontsize=10, color='black')
ax3.yaxis.set_ticks(np.arange(0, 19, 3))
ax3.set_ylim([0, 15])
ax3.tick_params(axis='y', labelcolor='black', labelsize=10)

bar_reduction = 100 * (rmsd_duacs_modul - rmsd_swot_modul) / rmsd_duacs_modul
ax3.text(0.21, 1.5, f'{bar_reduction:.0f}% RMSD \n reduction', ha='center', va='center',
         fontsize=9, fontweight='bold',
         bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='black', alpha = 0.9))

ax3_right.bar(x + 0.89, rmsd_duacs_dir, width * 1.2, color='darkcyan', zorder=2, label='DUACS', alpha=0.9)
ax3_right.bar(x + 1.205, rmsd_swot_dir, width * 1.2, color='skyblue', zorder=2, label='SWOT', alpha=0.9)
ax3_right.errorbar(x + 0.89, rmsd_duacs_dir, yerr=err_dir_duacs, fmt='o', color='yellowgreen', zorder=3, capsize=3)
ax3_right.errorbar(x + 1.205, rmsd_swot_dir, yerr=err_dir_swot, fmt='o', color='slateblue', zorder=3, capsize=3)

ax3_right.set_ylabel('RMSD [Â°]', fontsize=10, color='black')
ax3_right.set_ylim([0, 50])
ax3_right.tick_params(axis='y', labelcolor='black', labelsize=10)

bar_reduction = 100 * (rmsd_duacs_dir - rmsd_swot_dir) / rmsd_duacs_dir
ax3_right.text(1.04, 5, f'{bar_reduction:.0f}% RMSD \n reduction', ha='center', va='center',
               fontsize=9, fontweight='bold',
               bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='black', alpha = 0.9))

ax3.set_xticks([0.195, 1.045])
ax3.set_xticklabels(['Drifters vel. module', 'Drifters direction'], fontsize=10)

ax3.yaxis.grid(True)
ax3.legend(fontsize=9.7, bbox_to_anchor=(1.05, 1.16))

ax1.text(-0.07, 1.04, '(a)', fontsize=10.5, color='black', fontweight='bold')
ax2.text(-0.45, 15.6, '(b)', fontsize=10.5, color='black', fontweight='bold')
ax3.text(-0.55, 15.6, '(c)', fontsize=10.5, color='black', fontweight='bold')

plt.tight_layout()
# plt.savefig('rmsd_percent_improv_def_boxed.png', dpi=600)

# plt.savefig('../../grl_codes_to_publish_def_swotv2.0.1/figures_def/figure5_presentation.png',dpi=200, bbox_inches='tight')

plt.show()
```
